```python
def Gradient_descent(om,ob,l,points):
    mgrad = 0
    bgrad = 0
    n=len(points)
    for i in range(n):
        x=points.iloc[i].YearsExperience
        y=points.iloc[i].Salary
        mgrad += -x*(y-(om*x + ob))
        bgrad += -(y-(om*x + ob))
    
    m =om -  (2/n)*l*mgrad
    b =ob -  (2/n)*l*bgrad
    return m,b


    
        
        
    
    
```

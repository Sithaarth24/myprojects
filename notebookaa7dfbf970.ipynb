```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = pd.read_csv('/kaggle/input/salary-data-prediction/Salary_Data.csv')
print(data)
```

```python
plt.scatter(data.YearsExperience,data.Salary)
plt.show()
```

```python
def Gradient_descent(om,ob,l,points):
    mgrad = 0
    bgrad = 0
    n=len(points)
    for i in range(n):
        x=points.iloc[i].YearsExperience
        y=points.iloc[i].Salary
        mgrad += -x*(y-(om*x + ob))
        bgrad += -(y-(om*x + ob))
    
    m =om -  (2/n)*l*mgrad
    b =ob -  (2/n)*l*bgrad
    return m,b


    
        
        
    
    
```
